```python
# Import dependencies

import pandas as pd
```

```python
# Set the file path and use Pandas to read the CSV File

purchase_data_file = "Resources/purchase_data.csv"

purchase_data = pd.read_csv(purchase_data_file)
```

```python
# View the head of the data

purchase_data.head()
```

```python
# CLEAN THE DATA

# Drop duplicates and null values (there are none)

# Print a data frame checking the amount of unique player SNs (screen names)

Total_players = len(purchase_data['SN'].drop_duplicates())

Total_players_df = pd.DataFrame({'Total Players': [Total_players]})

#purchase_data.isnull().sum()

Total_players_df
```

```python
# Drop duplicate items to analyze each unique item

Unique_items = len(purchase_data['Item ID'].drop_duplicates())

# Average_price

Average_price = purchase_data['Price'].sum() / purchase_data['Purchase ID'].count()

# Number_of_purchases

Number_of_purchases = purchase_data['Purchase ID'].count()

# Total_revenue

Total_revenue = purchase_data['Price'].sum()

# Put the Item Analysis Data into a Pandas Dataframe

Purchase_analysis_df = pd.DataFrame({'Unique Items': [Unique_items],
                                    'Average Price': [Average_price],
                                    'Number of Purchases': [Number_of_purchases],
                                    'Total Revenue': [Total_revenue]})

Purchase_analysis_df['Average Price'] = Purchase_analysis_df['Average Price'].map('${:.2F}'.format)

Purchase_analysis_df['Total Revenue'] = Purchase_analysis_df['Total Revenue'].map('${:.2F}'.format)
```

```python
# Print the purchase_analysis dataframe
Purchase_analysis_df
```

```python
#Drop the duplicates and place in a new dataframe

No_duplicates = purchase_data.drop_duplicates('SN')

No_duplicates.head()
```

```python
# Group the dataframe by gender

Grouped_by_gender_df = No_duplicates.groupby(['Gender'])

# Select the SN Column

Only_SN = Grouped_by_gender_df['SN']

# Count the players by gender

Total_players_gender = Only_SN.count()

# Calculate percentages for each gender

Percentage_players = Total_players_gender / Total_players

# Place gender data in a dataframe and print

Gender_df = pd.DataFrame({'Total Players': Total_players_gender,
                         'Percentage Players': Percentage_players})

Gender_df['Percentage Players'] = Gender_df['Percentage Players'].astype(float).map('{:.2%}'.format)

Gender_df

# CHECK

# Total_players = Gender_df['Total Players'].sum()

# Total_players
```

```python
# Purchasing Analysis (Gender)

# Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender

Purchase_gender_df = purchase_data.groupby(['Gender'])

Purchase_count = Purchase_gender_df['Purchase ID'].count()

Avg_purchase_price = Purchase_gender_df['Price'].sum() / Purchase_count

Total_purchase_value = Purchase_gender_df['Price'].sum()

Avg_total_per_person = Purchase_gender_df['Price'].sum() / Total_players_gender


# Create a summary data frame to hold the results

# Place results in a dataframe

Purchase_summary_df = pd.DataFrame({'Purchase Count': Purchase_count,
                                   'Average Purchase Price': Avg_purchase_price,
                                   'Total Purchase Value': Total_purchase_value,
                                   'Avg Total Purchase Per Person': Avg_total_per_person})


# Give the displayed data cleaner formatting

# These values must be formatted using the map function to be properly placed in the dataframe

Purchase_summary_df['Average Purchase Price'] = Purchase_summary_df['Average Purchase Price'].astype(float).map('${:.2f}'.format)

Purchase_summary_df['Total Purchase Value'] = Purchase_summary_df['Total Purchase Value'].astype(float).map('${:.2f}'.format)

Purchase_summary_df['Avg Total Purchase Per Person'] = Purchase_summary_df['Avg Total Purchase Per Person'].astype(float).map('${:.2f}'.format)

# Display the summary data frame

Purchase_summary_df
```

```python
# Age Demographics


# Establish bins for ages

age_bins = [0, 9, 14, 19, 24, 29, 34, 39, 200]

age_labels = ['<9', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']

# Categorize the existing players using the age bins. Use pd.cut()

No_duplicates['Age Group'] = pd.cut(No_duplicates['Age'], age_bins, labels=age_labels)

# Calculate the numbers and percentages by age group

age_groups = No_duplicates.groupby('Age Group')

age_group_counts = age_groups['Age Group'].count()

Percent_age_group = age_group_counts / Total_players * 100

# Create a summary data frame to hold the results

Age_demo_df = pd.DataFrame({'Total Count': age_group_counts,
                           'Percentage of Players': Percent_age_group})

# Round the percentage column to two decimal points

Age_demo_df['Percentage of Players'] = Age_demo_df['Percentage of Players'].astype(float).map('{:.2f}%'.format)

# Display Age Demographics Table

Age_demo_df
```

```python
# Purchasing Analysis (Age)


# Bin the purchase_data data frame by age

age_bins = [0, 9, 14, 19, 24, 29, 34, 39, 200]

age_labels = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']

# Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below
purchase_data['Age Group'] = pd.cut(purchase_data['Age'], age_bins, labels=age_labels)

age_groups = purchase_data.groupby('Age Group')

age_group_counts = age_groups['Age Group'].count()

total_purchases = age_groups['Price'].sum()

avg_purchase_price = total_purchases / age_group_counts

player_ages = purchase_data.groupby('Age Group')

player_ages_counts = age_groups['Age Group'].count()

avg_purchase_total_per_person = total_purchases / age_groups['SN'].count()

# Create a summary data frame to hold the results

purchase_analysis_age_df = pd.DataFrame({'Purchase Count': age_group_counts,
                                        'Average Purchase Price': avg_purchase_price,
                                        'Total Purchase Value': total_purchases,
                                        'Avg Total Purchase per Person': avg_purchase_total_per_person})

# Give the displayed data cleaner formatting

purchase_analysis_age_df['Average Purchase Price'] = purchase_analysis_age_df['Average Purchase Price'].astype(float).map('${:.2f}'.format)

purchase_analysis_age_df['Total Purchase Value'] = purchase_analysis_age_df['Total Purchase Value'].astype(float).map('${:.2f}'.format)

purchase_analysis_age_df['Avg Total Purchase per Person'] = purchase_analysis_age_df['Avg Total Purchase per Person'].astype(float).map('${:.2f}'.format)

# Display the summary data frame
purchase_analysis_age_df
```

```python
# Top Spenders


# Run basic calculations to obtain the results in the table below
purchases_by_SN = purchase_data.groupby('SN')

SN_count_purchases = purchases_by_SN['SN'].count()

SN_total_purchases = purchases_by_SN['Price'].sum()

SN_avg_purchase_price = SN_total_purchases / SN_count_purchases

# Create a summary data frame to hold the results

Top_spenders_df = pd.DataFrame({'Purchase Count': SN_count_purchases,
                              'Average Purchase Price': SN_avg_purchase_price,
                               'Total Purchase Value': SN_total_purchases})

#Top_spenders_df

# Sort the total purchase value column in descending order

SN_total_purchases_sorted = Top_spenders_df.sort_values('Total Purchase Value', ascending=False)

# Give the displayed data cleaner formatting

SN_total_purchases_sorted['Total Purchase Value'] = SN_total_purchases_sorted['Total Purchase Value'].astype(float).map('${:.2f}'.format)

SN_total_purchases_sorted['Average Purchase Price'] = SN_total_purchases_sorted['Average Purchase Price'].astype(float).map('${:.2f}'.format)

# Display a preview of the summary data frame

SN_total_purchases_sorted.head(5)
```

```python
# Most Popular Items

# Retrieve the Item ID, Item Name, and Item Price columns

purchase_data_copy_2 = purchase_data.copy()

# Group by Item ID and Item Name.

purchase_data_group_by2 = purchase_data_copy_2.groupby(['Item ID', 'Item Name'])

# Perform calculations to obtain purchase count, item price, and total purchase value

purchase_count = purchase_data_group_by2['Item ID'].count()

item_price = purchase_data_group_by2['Price'].sum() / purchase_count

total_purchase_value = purchase_count * item_price

# Create a summary data frame to hold the results

Most_popular_df = pd.DataFrame({'Purchase Count': purchase_count,
                               'Item Price': item_price,
                               'Total Purchase Value': total_purchase_value})

# Sort the purchase count column in descending order

Most_popular_df = Most_popular_df.sort_values('Purchase Count', ascending=False)

# Give the displayed data cleaner formatting

Most_popular_df['Item Price'] = Most_popular_df['Item Price'].astype(float).map('${:.2f}'.format)

Most_popular_df['Total Purchase Value'] = Most_popular_df['Total Purchase Value'].astype(float).map('${:.2f}'.format)

# Display a preview of the summary data frame

Most_popular_df.head(5)
```

```python
# Most Profitable Items

# Retrieve the Item ID, Item Name, and Item Price columns

purchase_data_copy_2 = purchase_data.copy()

# Group by Item ID and Item Name.

purchase_data_group_by2 = purchase_data_copy_2.groupby(['Item ID', 'Item Name'])

# Perform calculations to obtain purchase count, item price, and total purchase value

purchase_count = purchase_data_group_by2['Item ID'].count()

item_price = purchase_data_group_by2['Price'].sum() / purchase_count

total_purchase_value = purchase_count * item_price

# Display a preview of the summary data frame

Most_profitable_df = pd.DataFrame({'Purchase Count': purchase_count,
                               'Item Price': item_price,
                               'Total Purchase Value': total_purchase_value})

# Sort the purchase count column in descending order

Most_profitable_df = Most_profitable_df.sort_values('Total Purchase Value', ascending=False)

# Give the displayed data cleaner formatting

Most_profitable_df['Item Price'] = Most_profitable_df['Item Price'].astype(float).map('${:.2f}'.format)

Most_profitable_df['Total Purchase Value'] = Most_profitable_df['Total Purchase Value'].astype(float).map('${:.2f}'.format)


# Create a summary data frame to hold the results

Most_profitable_df.head(5)
```
