```python
# Purchasing Analysis (Gender)

# Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender
Purchase_gender_df = purchase_data.groupby(['Gender'])
Purchase_count = Purchase_gender_df['Purchase ID'].count()
Avg_purchase_price = Purchase_gender_df['Price'].sum() / Purchase_count
Total_purchase_value = Purchase_gender_df['Price'].sum()
Avg_total_per_person = Purchase_gender_df['Price'].sum() / Total_players_gender

Purchase_summary_df = pd.DataFrame({'Purchase Count': Purchase_count,
                                   'Average Purchase Price': Avg_purchase_price,
                                   'Total Purchase Value': Total_purchase_value,
                                   'Avg Total Purchase Per Person': Avg_total_per_person})
Purchase_summary_df['Average Purchase Price'] = Purchase_summary_df['Average Purchase Price'].astype(float).map('${:.2f}'.format)
Purchase_summary_df['Total Purchase Value'] = Purchase_summary_df['Total Purchase Value'].astype(float).map('${:.2f}'.format)
Purchase_summary_df['Avg Total Purchase Per Person'] = Purchase_summary_df['Avg Total Purchase Per Person'].astype(float).map('${:.2f}'.format)
Purchase_summary_df

# Create a summary data frame to hold the results
# Optional: give the displayed data cleaner formatting
# Display the summary data frame
```

```python
# Age Demographics
# Establish bins for ages
# Categorize the existing players using the age bins. Hint: use pd.cut()
# Calculate the numbers and percentages by age group
# Create a summary data frame to hold the results
# Optional: round the percentage column to two decimal points
# Display Age Demographics Table

age_bins = [0, 9, 14, 19, 24, 29, 34, 39, 200]
age_labels = ['<9', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']
No_duplicates['Age Group'] = pd.cut(No_duplicates['Age'], age_bins, labels=age_labels)
age_groups = No_duplicates.groupby('Age Group')
age_group_counts = age_groups['Age Group'].count()
Percent_age_group = age_group_counts / Total_players * 100

# purchase_data
Age_demo_df = pd.DataFrame({'Total Count': age_group_counts,
                           'Percentage of Players': Percent_age_group})
Age_demo_df['Percentage of Players'] = Age_demo_df['Percentage of Players'].astype(float).map('{:.2f}%'.format)
Age_demo_df
# Age_demo_df
```

```python
# Purchasing Analysis (Age)
# Bin the purchase_data data frame by age
# Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below
# Create a summary data frame to hold the results
# Optional: give the displayed data cleaner formatting
# Display the summary data frame
age_bins = [0, 9, 14, 19, 24, 29, 34, 39, 200]
age_labels = ['<9', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']
purchase_data['Age Group'] = pd.cut(purchase_data['Age'], age_bins, labels=age_labels)
age_groups = purchase_data.groupby('Age Group')
age_group_counts = age_groups['Age Group'].count()
total_purchases = age_groups['Price'].sum()
avg_purchase_price = total_purchases / age_group_counts
player_ages = purchase_data.groupby('Age Group')
player_ages_counts = age_groups['Age Group'].count()
avg_purchase_total_per_person = total_purchases / player_ages_counts

purchase_analysis_age_df = pd.DataFrame({'Purchase Count': age_group_counts,
                                        'Average Purchase Price': avg_purchase_price,
                                        'Total Purchase Value': total_purchases,
                                        'Avg Total Purchase per Person': avg_purchase_total_per_person})
purchase_analysis_age_df['Average Purchase Price'] = purchase_analysis_age_df['Average Purchase Price'].astype(float).map('${:.2f}'.format)
purchase_analysis_age_df['Total Purchase Value'] = purchase_analysis_age_df['Total Purchase Value'].astype(float).map('${:.2f}'.format)
purchase_analysis_age_df['Avg Total Purchase per Person'] = purchase_analysis_age_df['Avg Total Purchase per Person'].astype(float).map('${:.2f}'.format)

purchase_analysis_age_df
```

```python
# Top Spenders
# Run basic calculations to obtain the results in the table below
# Create a summary data frame to hold the results
# Sort the total purchase value column in descending order
# Optional: give the displayed data cleaner formatting
# Display a preview of the summary data frame
purchases_by_SN = purchase_data.groupby('SN')
SN_count_purchases = purchases_by_SN['SN'].count()
SN_total_purchases = purchases_by_SN['Price'].sum()
SN_avg_purchase_price = SN_total_purchases / SN_count_purchases

Top_spenders_df = pd.DataFrame({'Purchase Total': SN_total_purchases,
                               'Total Purchases': SN_count_purchases,
                              'Avg Purchase Price': SN_avg_purchase_price})
Top_spenders_df
SN_total_purchases_sorted = Top_spenders_df.sort_values('Purchase Total', ascending=False)
SN_total_purchases_sorted['Purchase Total'] = SN_total_purchases_sorted['Purchase Total'].astype(float).map('${:.2f}'.format)
SN_total_purchases_sorted['Avg Purchase Price'] = SN_total_purchases_sorted['Avg Purchase Price'].astype(float).map('${:.2f}'.format)

SN_total_purchases_sorted.head(5)
```
